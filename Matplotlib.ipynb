# Generating some graphs using Matplolib

In this section, we will explore ways to visualize data using Matplotlib

```python
# imports & loading data into DataFrame

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv('adult.csv')
df.replace('?', np.nan, inplace = True)
df.dropna(inplace = True)
```

### I. Pairwise data

1. Bar

```python
fig, ax = plt.subplots(figsize=(12, 6)); # resizes the plot
labels = plt.xticks(rotation='vertical') # rotating labels to be vertical

x = df['marital.status'].unique() # getting all unique values in the column
y = []

for ms in x:
    y.append(df[df['marital.status'] == ms].shape[0]) # getting the number of instances for each status

ax.bar(x, y)
plt.xlabel('marital status')
plt.title('Distribution of Marital Status')
plt.show()
```

2. Stem

```python
fig, ax = plt.subplots(figsize=(12, 6)); # resizes the plot
x = df['workclass'].unique() # getting all the unique values in the column
y = []

for workclass in x:
    y.append(df[df['workclass'] == workclass].shape[0]) # getting the number of instances for each class

ax.stem(x, y)
plt.xlabel('Workclass')
plt.title('Distribution of Workclass')

plt.show()
```

### Statistics

1. Histogram

```python
fig, ax = plt.subplots(figsize=(12, 6)); # resizes the plot
x = df['age']

ax.hist(x, bins=10, linewidth=0.5, edgecolor="white")
plt.xlabel('age')
plt.title('Distribution of Age')
ax.set(xlim=(5, 95))

plt.show()
```

2. Pie chart

```python
fig, ax = plt.subplots(figsize=(12, 6)); # resizes the plot
x = [df[df['sex'] == 'Female'].shape[0], df[df['sex'] == 'Male'].shape[0]] # creating a list with the number of Females and Males respectively.
labels = ['Female', 'Male'] # creating labels for the legend

ax.pie(x)
plt.title('Distribution of Sex')
plt.legend(labels, loc="best") # displaying the legend; the location is determined automatically using 'best'

plt.show()
```
